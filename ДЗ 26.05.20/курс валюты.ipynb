```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/it_new.csv')
df.head(10)
```

```python
import requests  
response = requests.get('https://www.cbr-xml-daily.ru/daily_json.js') 
current = response.json()
current['Valute'].keys()
```

### Задание

Определите заработную плату в ₽ по текущему курсу.

```python
df['curenc'] = df['curenc'].replace({'zł':'PLN','Kč':'CZK','Ft':'HUF','$':'USD','lev':'BGN','€':'EUR'})
def to_RUB(row):
    if row['curenc'] != '₽':
        return round(row['col']*current['Valute'][row['curenc']]['Value'], 2) 
    else:
        return row['col']  
```

```python
df['З/п в ₽ по текущему курсу'] = df.apply(to_RUB, axis = 1)
df
```
